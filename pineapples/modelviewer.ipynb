```python
%matplotlib inline
import numpy as np
from array import array
import matplotlib
import matplotlib.pyplot as plt



nx=300
nz=100
dx=0.025
dz=0.025

xtick=1.0
ztick=0.5

def scale_xaxis(number):
    return(number*dx)
def scale_zaxis(number):
    return(number*dz)
def my_xform(x,pos):
    return '%.1f'%scale_xaxis(x)
def my_zform(z,pos):
    return '%.1f'%scale_zaxis(z)

data=np.fromfile('marmousi_vp',dtype=np.float32)

vpmodel=np.reshape(data,(nz,nx))

data=np.fromfile('marmousi_vs',dtype=np.float32)

vsmodel=np.reshape(data,(nz,nx))

data=np.fromfile('marmousi_rho',dtype=np.float32)

rhomodel=np.reshape(data,(nz,nx))

data=np.fromfile('marmousi_vp_sm',dtype=np.float32)

vpSMmodel=np.reshape(data,(nz,nx))

data=np.fromfile('marmousi_vs_sm',dtype=np.float32)

vsSMmodel=np.reshape(data,(nz,nx))

data=np.fromfile('marmousi_rho_sm',dtype=np.float32)

rhoSMmodel=np.reshape(data,(nz,nx))


```

```python

fig=plt.figure(figsize=(18,8))
plt.subplot(231)
ax=plt.gca()
ax.set_ylabel("Depth [km]")
ax.set_xlabel("Horizontal [km]")
#plt.yticks(np.arange(0,-nz*dz,-0.5))
#plt.xticks(np.arange(0,nx*dx,1.0))
plt.imshow(vpmodel,aspect='auto')
ax.autoscale(False)
ax.set_title('Vp model')
ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(int(xtick/dx)))
ax.xaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(my_xform))
ax.yaxis.set_major_locator(matplotlib.ticker.MultipleLocator(int(ztick/dz)))
ax.yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(my_zform))
plt.colorbar()

plt.subplot(232)
ax=plt.gca()
ax.set_ylabel("Depth [km]")
ax.set_xlabel("Horizontal [km]")
#plt.yticks(np.arange(0,-nz*dz,-0.5))
#plt.xticks(np.arange(0,nx*dx,1.0))
plt.imshow(vsmodel,aspect='auto')
ax.autoscale(False)
ax.set_title('Vs model')
ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(int(xtick/dx)))
ax.xaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(my_xform))
ax.yaxis.set_major_locator(matplotlib.ticker.MultipleLocator(int(ztick/dz)))
ax.yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(my_zform))
plt.colorbar()

plt.subplot(233)
ax=plt.gca()
ax.set_ylabel("Depth [km]")
ax.set_xlabel("Horizontal [km]")
#plt.yticks(np.arange(0,-nz*dz,-0.5))
#plt.xticks(np.arange(0,nx*dx,1.0))
plt.imshow(rhomodel,aspect='auto')
ax.autoscale(False)
ax.set_title('Rho model')
ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(int(xtick/dx)))
ax.xaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(my_xform))
ax.yaxis.set_major_locator(matplotlib.ticker.MultipleLocator(int(ztick/dz)))
ax.yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(my_zform))
plt.colorbar()

plt.subplot(234)
ax=plt.gca()
ax.set_ylabel("Depth [km]")
ax.set_xlabel("Horizontal [km]")
#plt.yticks(np.arange(0,-nz*dz,-0.5))
#plt.xticks(np.arange(0,nx*dx,1.0))
plt.imshow(vpSMmodel,aspect='auto')
ax.autoscale(False)
ax.set_title('Vp model')
ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(int(xtick/dx)))
ax.xaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(my_xform))
ax.yaxis.set_major_locator(matplotlib.ticker.MultipleLocator(int(ztick/dz)))
ax.yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(my_zform))
plt.colorbar()

plt.subplot(235)
ax=plt.gca()
ax.set_ylabel("Depth [km]")
ax.set_xlabel("Horizontal [km]")
#plt.yticks(np.arange(0,-nz*dz,-0.5))
#plt.xticks(np.arange(0,nx*dx,1.0))
plt.imshow(vsSMmodel,aspect='auto')
ax.autoscale(False)
ax.set_title('Vs model')
ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(int(xtick/dx)))
ax.xaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(my_xform))
ax.yaxis.set_major_locator(matplotlib.ticker.MultipleLocator(int(ztick/dz)))
ax.yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(my_zform))
plt.colorbar()

plt.subplot(236)
ax=plt.gca()
ax.set_ylabel("Depth [km]")
ax.set_xlabel("Horizontal [km]")
#plt.yticks(np.arange(0,-nz*dz,-0.5))
#plt.xticks(np.arange(0,nx*dx,1.0))
plt.imshow(rhoSMmodel,aspect='auto')
ax.autoscale(False)
ax.set_title('Rho model')
ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(int(xtick/dx)))
ax.xaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(my_xform))
ax.yaxis.set_major_locator(matplotlib.ticker.MultipleLocator(int(ztick/dz)))
ax.yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(my_zform))
plt.colorbar()
```
